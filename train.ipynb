```python
import os
import json
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset
import torchvision.transforms as transforms
from torchvision.transforms import Resize, CenterCrop, Normalize
from torchvision.models import resnet50
from torchvision.transforms.functional import pil_to_tensor
import cv2
from PIL import Image

# 학습 데이터 폴더 경로
data_folder = "train2/images"

# GPU 설정
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
num_gpus = 1  # 하나의 GPU만 사용

# 하이퍼파라미터
num_epochs = 10
batch_size_per_gpu = 16

class CustomDataset(Dataset):
    def __init__(self, data_folder, target_size=(256, 256)):
        self.image_paths = [os.path.join(data_folder, img_name) for img_name in os.listdir(data_folder) if img_name.endswith(".jpg")]
        self.labels_folder = os.path.join(data_folder, "../labels")
        self.target_size = target_size
        self.transform = transforms.Compose([
            # 리사이즈는 하지 않음
#             CenterCrop(target_size),
            transforms.ToTensor(),
            transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
        ])

    def load_label(self, image_path):
        image_name = os.path.basename(image_path)
        label_path = os.path.join(self.labels_folder, image_name.replace(".jpg", ".json"))
        try:
            with open(label_path, "r") as f:
                label = json.loads(f.read())["annotations"]["drawing"][0]
            return [label]  # 리스트의 리스트 형태로 반환
        except:
            return []


    def __len__(self):
        return len(self.image_paths)

    def resize_labels(self, label, original_size, target_size):
        # 리사이즈 이전과 이후의 크기 비율 계산
        ratio_x = target_size[0] / original_size[0]
        ratio_y = target_size[1] / original_size[1]
        
        # 상대 좌표로 변환
        for i in range(len(label)):
            label[i] = [(x * ratio_x, y * ratio_y) for x, y in label[i]]
        return label

    def __getitem__(self, idx):
        image_path = self.image_paths[idx]
        image = cv2.imread(image_path)
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

        original_size = (image.shape[1], image.shape[0])  # 이미지의 원래 크기 저장

        label = self.load_label(image_path)

        # 이미지를 target_size로 리사이즈
        image = self.transform(Image.fromarray(image))
        
        # 라벨 좌표를 리사이즈에 맞게 변환
        label = self.resize_labels(label, original_size, self.target_size)

        return image, label

# 모델 정의
class ObjectDetectionModel(nn.Module):
    def __init__(self, num_classes):
        super(ObjectDetectionModel, self).__init__()
        self.resnet = resnet50(pretrained=True)
        in_features = self.resnet.fc.in_features
        self.resnet.fc = nn.Linear(in_features, num_classes)

    def forward(self, x):
        return self.resnet(x)

# 데이터 로더 생성
dataset = CustomDataset(data_folder)
train_loader = DataLoader(dataset, batch_size=batch_size_per_gpu*num_gpus, shuffle=True, num_workers=num_gpus)

# 모델 및 손실 함수, 옵티마이저 정의
model = ObjectDetectionModel(num_classes=1).to(device)
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)

# 학습 루프
for epoch in range(num_epochs):
    model.train()
    total_loss = 0

    for images, labels in train_loader:
        images = images.to(device)
        labels = labels.to(device)

        optimizer.zero_grad()

        outputs = model(images)
        loss = criterion(outputs.squeeze(), labels.float())
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print(f"Epoch {epoch+1}/{num_epochs}, Loss: {total_loss / len(train_loader)}")

print("Training completed!")
```

```python
import os
import json
import torch
import torch.nn as nn
import torch.optim as optim
import torchvision.transforms as transforms
from torch.utils.data import DataLoader, Dataset
from torchvision.datasets import ImageFolder
from torch.utils.data.dataset import random_split
from torchvision.models import resnet50

# 하이퍼파라미터 설정
batch_size = 32
learning_rate = 0.001
num_epochs = 10
num_classes = 1  # 객체의 클래스 개수 (하나의 객체만 인식하는 경우는 1)
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")

# 데이터셋 클래스 정의
class CustomDataset(Dataset):
    def __init__(self, image_folder, label_folder, transform=None):
        self.image_folder = image_folder
        self.label_folder = label_folder
        self.transform = transform
        self.image_list = os.listdir(image_folder)

    def __len__(self):
        return len(self.image_list)

    def __getitem__(self, idx):
        image_name = self.image_list[idx]
        image_path = os.path.join(self.image_folder, image_name)
        image = Image.open(image_path).convert('RGB')
        if self.transform:
            image = self.transform(image)

        label_file = os.path.join(self.label_folder, image_name.replace(".jpg", ".json"))
        with open(label_file, "r") as f:
            label_data = json.loads(f.read())
            try:
                label = label_data["annotations"]["drawing"][0]  # JSON 파일에 label이라는 key로 라벨이 저장되어 있다고 가정
                print(label)
            except:
                label = []

        return image, label

# 데이터 로딩 및 전처리
transform = transforms.Compose([
    transforms.Resize((224, 224)),  # 이미지 크기 조정
    transforms.ToTensor(),          # 이미지를 Tensor로 변환
    transforms.Normalize((0.485, 0.456, 0.406), (0.229, 0.224, 0.225))  # 이미지 정규화
])

image_folder = "train2/images"
label_folder = "train2/labels"

dataset = CustomDataset(image_folder=image_folder, label_folder=label_folder, transform=transform)

# 데이터 분할 (학습 데이터와 검증 데이터로 나눔)
train_size = int(0.8 * len(dataset))
val_size = len(dataset) - train_size
train_dataset, val_dataset = random_split(dataset, [train_size, val_size])

# DataLoader 생성
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=batch_size)

# 모델 구성 (예시로 ResNet-50 사용)
model = resnet50(pretrained=True)
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, num_classes)
model.to(device)

# 손실 함수 및 옵티마이저 설정
criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=learning_rate)

# 학습
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for images, labels in train_loader:
        images = [img.to(device) for img in images]  # Move each image to the device
        labels = labels.to(device)  # Move labels to the device
        optimizer.zero_grad()
        outputs = model(images[0])  # Assuming only one image per batch
        loss = criterion(outputs, labels.float().unsqueeze(1))
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch [{epoch + 1}/{num_epochs}] - Loss: {running_loss / len(train_loader)}")

# 평가
model.eval()
with torch.no_grad():
    correct = 0
    total = 0
    for images, labels in val_loader:
        images, labels = images.to(device), labels.to(device)
        outputs = model(images)
        predicted = (outputs >= 0.5).float()
        total += labels.size(0)
        correct += (predicted == labels.float().unsqueeze(1)).sum().item()
    print(f"Validation Accuracy: {100 * correct / total}%")
```

```python
def convert_to_yolo_format(image_width, image_height, x1, x2, y1, y2):
    # 이미지 중심 좌표와 너비, 높이 계산
    center_x = (x1 + x2) / 2
    center_y = (y1 + y2) / 2
    width = x2 - x1
    height = y2 - y1

    # YOLO 형식으로 변환
    yolo_center_x = center_x / image_width
    yolo_center_y = center_y / image_height
    yolo_width = width / image_width
    yolo_height = height / image_height

    return 0, yolo_center_x, yolo_center_y, yolo_width, yolo_height

# 이미지의 가로와 세로 크기
iWidth = 1920
iHeight = 1080

import os
import glob
import json

jsonFiles = glob.glob(os.path.join('train2/labels_json/*'))
# print(jsonFiles)

for file in jsonFiles:
    with open(os.path.join(file), 'r') as f:
        f = json.loads(f.read())
        try:
            x1 = f["annotations"]["drawing"][0][0][0]
            x2 = f["annotations"]["drawing"][0][1][0]
            y1 = f["annotations"]["drawing"][0][0][1]
            y2 = f["annotations"]["drawing"][0][2][1]
    #         print(x1, x2, y1, y2)

            r = convert_to_yolo_format(iWidth, iHeight, x1, x2, y1, y2)
            print(r)
            fo = open(os.path.join('./train2/labels/', f'{os.path.basename(file)[:-5]}.txt'), "w")
            fo.write(str(r))
            fo.close()
                
        except Exception as e:
            fo = open(os.path.join('./train2/labels/', f'{os.path.basename(file)[:-5]}.txt'), "w")
            fo.write("")
            fo.close()

# YOLO 형식으로 변환
#yolo_center_x, yolo_center_y, yolo_width, yolo_height = convert_to_yolo_format(image_width, image_height, x1, x2, y1, y2)

# 변환된 결과 출력
#print("YOLO format: center_x={}, center_y={}, width={}, height={}".format(yolo_center_x, yolo_center_y, yolo_width, yolo_height))
```

```python
import os
import shutil

def copy_files(source_folder, destination_folder, num_files=1000):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)

    # Get a list of files in the source folder
    file_list = os.listdir(source_folder)

    # Copy the first `num_files` files to the destination folder
    for i, filename in enumerate(file_list):
        if i >= num_files:
            break
        source_path = os.path.join(source_folder, filename)
        source_path2 = os.path.join(f"{source_folder}/../../labels/train", f"{filename[:-4]}.txt")
        destination_path = os.path.join(destination_folder, filename)
        destination_path2 = os.path.join(f"{destination_folder}/../../labels/val", f"{filename[:-4]}.txt")
        shutil.copy2(source_path, destination_path)
        shutil.copy2(source_path2, destination_path2)
    

if __name__ == "__main__":
    source_folder = "train2/images/train"  # 원본 폴더의 경로를 입력하세요
    destination_folder = "train2/images/val"  # 복사할 폴더의 경로를 입력하세요
    num_files_to_copy = 300

    copy_files(source_folder, destination_folder, num_files_to_copy)
```
